## **Automating Tasks**

Generative AI can help automate many data-related tasks. Let's see some examples.

### **Data Labelling**

Say we want to train a model that classifies tweets as happy, neutral, or sad. We have 20,000 tweets of training data, but they are unlabelled. What can we do?

We have a few traditional options:


1.   Manually label each tweet
2.   Outsource labelling (ex. Amazon Mechanical Turk)

Option 1 can be time intensive and option 2 can be inaccurate for more difficult tasks.

Generative AI gives us option 3:
3. Allowing a Generative AI tool to label the data for us.
This could be expensive (depending on the tool), but is fast and effective.

We might ask, why do we need to train a model at all if Generative AI can do it for us? The reason is that using a large AI model to do this all the time in production would likely be more expensive than needed.

Each of these options have their own tradeoffs and use cases depending on the situation, but Generative AI now gives us another option.

If you're interested in learning more, Dr. Daniel Ringel is doing some cool research related! [Check it out here](http://www.synthetic-experts.ai/)!



### **Data Augmentation**

Using the same example of training a model that classifies tweets as happy, neutral, or sad. However, what if we only had 200 tweets of data. This might not be enough for us to feel confident in.

We could give Generative AI some examples of the tweets we're looking for and ask it to generate thousands of more tweets.

This probably isn't as good as actually collecting thousands of more tweets, but it could give us a quick, easy solution.

### **Data Masking**
Sticking with the same example of tweets, what if our tweets had sensitive data such as names, addresses, phone numbers etc. This data might not be important to what we are trying to predict, but we wouldn't want this information getting leaked if we decide to post our data online.

We can use Generative AI to annonymize our data, prompting it to reword the data with different names, addresses, phone numbers etc.

Dr. Ringel also discusses this concept of "Synthetic Twins" in his research. [Check it out here](https://www.synthetic-experts.ai/)!

## **Support with Writing Code**

We'll look at an example of how we can use AI to do a few very basic analyses on a UNC Faculty Salaries Dataset. ([Source](https://uncdm.northcarolina.edu/salaries/index.php))

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/carolinadatascience/AI-Tools-Workshop/main/Salary%20Data%20Export.csv')
```

```python
df.head()
```

### **Example #1: Colab AI**

First let's test out Colab AI to see if it can do something really basic: finding the average annual base salary of all employees in this dataset.

```python
df['EMPLOYEE ANNUAL BASE SALARY'].mean()
```

Let's take it up a level. Let's try finding the average annual base salary broken down by departmnt. Note: You might get an error here.

```python
df.groupby('EMPLOYEE HOME DEPARTMENT')['EMPLOYEE ANNUAL BASE SALARY'].mean()
```

If you get an error, reword the prompt, to help the AI out. Let's also ask it to store this into a dataframe so that we can reuse it in our next step!

```python
average_salary_by_department = df.groupby('EMPLOYEE HOME DEPARTMENT')['EMPLOYEE ANNUAL BASE SALARY'].mean()
average_salary_by_department_df = pd.DataFrame(average_salary_by_department)
average_salary_by_department_df.head()
```

Now, let's display our findings looking at the top 10 average salaries by department. Try giving the graph a title and axis.

```python
average_salary_by_department_df.sort_values('EMPLOYEE ANNUAL BASE SALARY', ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 Paying Employee Home Departments')
plt.xlabel('Employee Home Department')
plt.ylabel('Average Salary')
plt.show()
```

**Overall, we can see that this was easy to use, but it still requires us to have Python knowledge so that we can troubleshoot when it goes wrong.**
